```python
seller_state = list(teste_sankey['source'].unique())
buyer_state = list(teste_sankey['target'].unique())
```

```python
# Set the default size to fit the documentation better.
size = dict(width=8000, height=8000)
nodes = {
    'seller_state': ProcessGroup(seller_state),
    'buyer_state': ProcessGroup(buyer_state),
}

ordering = [['seller_state'],['buyer_state']]
```

```python
sdd = SankeyDefinition(nodes, bundles, ordering)
weave(sdd, teste_sankey).to_widget()
```

```python

# The first argument is the dimension name -- for now we're using
# "process" to group by process ids. The second argument is a list
# of groups.
farms_with_other = Partition.Simple('process', seller_state)

# This is another partition.
customers_by_name = Partition.Simple('process', buyer_state)

# Update the ProcessGroup nodes to use the partitions
nodes['seller_state'].partition = farms_with_other
nodes['buyer_state'].partition = customers_by_name

# New Sankey!
weave(sdd, teste_sankey).to_widget(**size)
```
